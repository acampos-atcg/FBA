## Introducción

Como en cualquier otro organismo la glucosa es un sustrato esencial, el cual puede ser procesado a través de la vía de la glucólisis de Embden-Meyerhof-Parnas (EMP por sus siglas en inglés) para llevar a cabo sus dos principales funciones:

1) Cuando los aceptores terminales de electrones ya no están disponibles, la glicólisis provee todas la moléculas de ATP requeridas para la actividad celular. 

2) Los metabolitos intermediarios de la vía son precursores directos de muchas moléculas (Bar-Even, 2012).

*E. coli* utliza la enzima NAD+-dependiente gliceraldehído-3-fosfato deshidrogenasa (NAD+-GAPDH) codificada en el gen *gapA* (D'Alessio y Josse 1971) en la vía de la glicólisis. La actividad de dicha enzima es vital para el crecimiento del organismo (Charpentier y Branlant 1994), y la disponibilidad del cofactor NAD+ es esencial para su buen funcionamiento. Se ha reportado que cuando se sustituye el cofactor original por NADP+ (por medio del cambio la enzima, pero conservando el promotor de *gapA* u otro promotor constitutivo de un plásmido) en la vía de *E. coli*, es posible restaurar la actividad glucolítica de la bacteria aunque en menor cantidad que en la cepa silvestre (Centeno-Leijia, 2013).

A través del enfoque matemático conocido como análisis de balance de flujos (FBA por sus siglas en inglés), se puede analizar y calcular el flujo de metabolitos a través de una red metabólica, además de hacer predicciones de la tasa de crecimiento de un organismo, o bien, de la tasa producción de un metabolito biotecnológicamente relevante (Orth, 2010). Este enfoque utiliza una *función objetivo*, la cual será optimizada. Otra característica esencial de este tipo de modelado es que opera a través de constricciones. Para entender las constricciones, primero se describe cómo se representan las reacciones metabólicas en una matriz numérica (contiene los coeficientes estequiométricos de cada reacción). Estos valores determinan las constricciones del flujo de los metabolitos en la red, las cuales pueden ser representadas de dos maneras: a) ecuaciones que balancean las reacciones de entrada y de salida; b)desigualdades que imponen límites al sistema (Orth, 2010).

En el presente trabajo se modela a través de FBA el flujo de la vía que incluye al gen *gapA* y el cambio del cofactor original de la enzima *gapA*: NADP+. También se manipulan parámetros de otros metabolitos esenciales para el organismo, tales como el O2, CO2 y glucosa de acuerdo con trabajos experimentales previos que estudiaban la misma ruta, y el efecto del cambio de cofactor. Adicionalmente, se restringe el flujo de la enzima *gapA* y en todos los casos se busca hacer una comparación del crecimiento del organismo y por último, se grafica el cambio del flujo de la enzima vs el crecimiento máximo del individuo.

Para desarrollar el trabajo se hizo uso de la paquetería de código abierto: openCobra. Con ayuda de esto se manipulan los metabolitos, genes y reacciones de interés.<br>

### Definir fondo genético

Para interrogar la respuesta de E.coli en la via Embden-Meyerhof-Parnas (EMP) se modificó la reaccion catalizada por la enzima Gliceraldehido-3-fosfato-deshidrogenasa dependiente de NAD (GAPD). Dado que no se empleó una construcción genética en el modelo, se cambiaron los cofactores NAD y NADH por NADP y NADPH, respectivamente, simulando la reacción llevada a cabo por la GAPD, sólo que dependiente de NADP. El experimento in silico simula a aquél llevado a cabo in vitro con la deleción de gapA y la transformación con gapN (Centeno S., *et al*, 2013).

```python
#Cambio de cofactor
#Visualizacion de la reaccion de interes.
print("Reaccion sin modificar", gapd.reaction)

#Para hacer el cambio de cofactor se hace uso de dos funciones:
#(1) pop(). La cual recibe como parametro el metabolito que se desea remover, ademas de especificar la reaccion y el modelo. 
#Cabe destacar que la funcion lleva a cabo el balance de la reaccion.
#(2) add_metabolites(). Recibe como parametro el metabolito que se desea agregar a la reaccion, ademas de especificar el coeficiente de entrada y salida.

#(1)
gapd.pop(model.metabolites.get_by_id("nad_c"))
gapd.pop(model.metabolites.get_by_id("nadh_c"))
#(2)
gapd.add_metabolites({model.metabolites.get_by_id("nadp_c"): -1})
gapd.add_metabolites({model.metabolites.get_by_id("nadph_c"): 1})

#Visualizacion de la reaccion modificada.
print("Reaccion modificada", gapd.reaction)

#Simulacion FBA.
model.optimize()
anwmetch = model.solution.f
print 'Solucion de optimizacion:', anwmetch
#Modulo de despliegue del mapa utilizando escher
b = escher.Builder(map_name='iJO1366.Central metabolism',
                   reaction_data=model.solution.x_dict,
                   # Color y tamano cambiar de acuerdo a valores de flujo
                   reaction_styles=['color', 'size', 'abs', 'text'],
                   # cambiar colores
                   reaction_scale=[{'type': 'min', 'color': '#00ccff', 'size': 4},
                                   {'type': 'mean', 'color': '#0000dd', 'size': 20},
                                   {'type': 'max', 'color': '#ff0000', 'size': 40}],
                   # Esconder metabolitos secundarios
                   hide_secondary_metabolites=True)
b.display_in_notebook()
```

### Cambio en condiciones iniciales (Gluc_Ex, O2, CO2)

Cambio de las tasas de respiracion (O2), consumo de CO2 y consumo de glucosa. Los valores fueron tomados de experimentos in vitro y modificados para ajustar el modelo a las condiciones expuestas en Centeno S, 2013.

```python
#Se declaran las reacciones de intercambio celular: CO2, O2 y Glucosa.
co2 = model.reactions.get_by_id("EX_co2_e_")
o2 = model.reactions.get_by_id("EX_o2_e_")
glc = model.reactions.get_by_id("EX_glc_e_")

#Se modifican los limites maximos y minimos de cada reaccion. 
#Cada valor fue tomado de experimentos in vitro (Centeno S. et al, 2013) a partir de los resultados especificos para cada metabolito.
co2.upper_bound = 13.73 #Limite superior
o2.upper_bound = 13.75  #Limite inferior
glc.upper_bound = 5.11
co2.lower_bound = -14.05
o2.lower_bound = -14.27
glc.lower_bound = -5.37

#Visualizacion: Reaccion de interes y limites modificados (inferior < X < maximo)
print(co2.reaction)
print(co2.lower_bound, "< CO2 <", co2.upper_bound)
print(o2.reaction)
print(o2.lower_bound, "< O2 <", o2.upper_bound)
print(glc.reaction)
print(glc.lower_bound, "< GLC <", glc.upper_bound)

#Simulacion FBA.
model.optimize()
anwmetchparch = model.solution.f
print 'Solucion de optimizacion:', anwmetchparch
#Modulo de despliegue de la solucion utilizando escher
b = escher.Builder(map_name='iJO1366.Central metabolism',
                   reaction_data=model.solution.x_dict,
                   # Color y tamano cambiar de acuerdo a valores de flujo
                   reaction_styles=['color', 'size', 'abs', 'text'],
                   # cambiar colores
                   reaction_scale=[{'type': 'min', 'color': '#00ccff', 'size': 4},
                                   {'type': 'mean', 'color': '#0000dd', 'size': 20},
                                   {'type': 'max', 'color': '#ff0000', 'size': 40}],
                   # Esconder metabolitos secundarios
                   hide_secondary_metabolites=True)
b.display_in_notebook()
```

## Discusión y conclusiones

### Comparación de mu max

El valor obtenido a partir de la optimización de la función objetivo de biomasa se encuentra relacionado con la tasa de crecimiento de *E. coli* dado que el flujo de crecimiento se encuentra definido en términos de los requerimientos biosintéticos de la composición de biomasa. [Pramanik y Keasling, 1997]. Es posible observar que se obtiene un valor máximo de en la función objetivo cuando se corre la simulación con la reconstrucción metabólica de *E. coli* sin alterar los parámetros iniciales, es decir, sin limites el cambio de las tasas de respiracion (O2), consumo de CO2, consumo de glucosa y con los metabolitos canónicos en la reacción de glucólisis.

```python
print anwunmod
```

Posterior al cambio del uso de metabolitos se pudo observar un incremento ligero en el valor de la función objetivo, lo cual no coincide con el decremento en la tasa de crecimiento observado por Centeno-Leija *et. al* en una cepa con *knockout* de *gapA* suplementada con *gapN* de *Streptococcus mutants*, la cual codifica para una enzima GAPDH dependiente de NADP+. 

```python
print anwmetch
```

Esto puede atribuirse al hecho de que la definición del fondo génico utilizada en el presente reporte consiste únicamente del cambio de cofactor utilizado por *gapA* sin alterar los parámetros de actividad enzimática de ésta, los cuales se saben diferentes a los de *gapN*: Esta enzima posee una afinidad por el gliceraldehido 3-P 10 veces mayor que *gapA* y presenta una fuerte inhibición por sustrato y NADPH (Centeno S., *et al*, 2013).

Es posible observar también que el valor de la función objetivo disminuye cuando se acotan los límites superiores e inferiores de cambio de las tasas de respiracion (O2), consumo de CO2 y consumo de glucosa. Para esto se utilizaron los valores encontrados en el artículo de Centeno-Leija *et. al* más menos la desviación estándar reportada para definir los límites superiores e inferiores respectivamente.  Dado que las reacciones modificadas son reacciones de intercambio y por convención se encuentran escritas como reacciones de exporte (e.g. glc_D_e <=> ), el importe máximo de un metabolito (es decir, los valores reportados en el artículo) se describe como un flujo negativo en el límite inferior (Orth *et al*, 2010 material sup.).


```python
print anwparch
```

Se observó también un decremento en el valor de la función objetivo cuando se probó el cambio en los metabolitos junto con los valores acotados de los límites superiores e inferiores de las tasas previamente mencionadas. Este valor, sin embargo, fue ligeramente superior al 0.483 encontrado el la prueba con valores acotados sin el cambio de metabolitos. Esto coincide con el incremento ligero observado en la comparación sin acotar los limites superiores e inferiores con (0.93) y sin (0.91) cambio de metabolitos para la reacción de glicólisis.

```python
print anwmetchparch
```

Finalmente se realizó un *knockout* de la enzima al acotar los límites superior e inferior de la reacción en la que interviene a 0. Esto resultó en un decremento drástico en el valor de la función objetivo, el cual es mucho más pronunciado si se utiliza sólo la reconstrucción con las reacciones centrales de metabolismo de *E. coli* (modelo *core*).

```python
print anwmetchparchknock
print anwcoremetchparchknock
```

Esto nos indica que, pese a que la reacción resulta un paso determinante en la ruta de glicólisis de la cual depende la mayor parte del metabolismo de *E. coli*, existen en la bacteria rutas alternas que le permiten competir contra alguna falla en *gapA* y restaurar de manera limitada el flujo de carbono y la disponibilidad de elementos reductores (NADH/NADPH) y por tanto el aumento en biomasa. Si solo se modelan las reacciones nucleares del metabolismo es posible observar que el valor de la función objetivo es prácticamente 0, lo cual nos habla de la importancia de esta enzima para la supervivencia de la bacteria.

### Actividad de gapA vs mu max

```python
####Grafica 1 3D
x=[]
y=[]
z=[]
model = cobra.io.read_sbml_model('Ec_iAF1260_flux2.txt')
for i in np.arange(10,20,0.1):
    for p in np.arange(i,20,0.1):
        model.reactions.GAPD.lower_bound=i
        model.reactions.GAPD.upper_bound=p
        model.optimize()
        x.append(i)
        y.append(p)
        z.append(model.solution.f)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot3D(x,y,z)
ax.set_xlabel('GAPD lower bound')
ax.set_ylabel('GAPD upper bound')
ax.set_zlabel('Growth')
plt.show()
```

En las gráficas es posible observar que el crecimiento óptimo de la bacteria se dá cuando el flujo en la reacción mediada por GAPDH es cercano a 17 unidades.  Si el flujo a través de esta via sobrepasa las $\sim 17$ unidades por el aumento en el límite inferior (Gráfica 2), entonces el crecimiento se ve afectado probablemente debido a una descompensación en el el radio de metabolitos oxidantes y reductores. De igual manera, si el flujo de esta reacción es forzado a ser menor de $\sim 17$ a través de modificar el límite superior, el valor de la función objetivo disminuye posiblemente debido a que la producción de agentes reductores y el proceso de glucosa no es suficiente para solventar las necesidades de la bacteria en crecimiento (Gráfica 3).

## Perspectivas

FBA constituye un acercamiento útil, y debido a que sus fundamentos son simples de implementar, este enfoque ha sido utilizado en diversas áreas desde la psicología hasta la biología sintética y estudios genómicos (Orth, 2010). Es evidente la gran utilidad de este método, no sólo para fines biotecnológicos sino también para la resolución de problemas de investigación básica.

Desde el punto de vista de biología sintética, cuando se alteran los límites de ciertas reacciones, se pueden simular crecimientos en distintos medios de cultivo o incluso simular múltiples *knockouts*.  Es por esto que FBA ayuda a proveer de predicciones de cofactores importantes tales como el ATP, NADH o NADPH (Orth, 2010). 

Actualmente, todas las reconstrucciones metabólicas a escala genómica tienen reacciones faltantes, pero por medio de FBA es posible predecir cuáles son estas reacciones, haciendo comparaciones de crecimiento entre simulaciones *in silico* y experimentales. Sin embargo, como cualquier otro modelo, la validación biológica resulta sumamente importante para comprobar las predicciones arrojadas por el modelo. Estudio muestran que las tasa de crecimiento de *E. coli* en presencia de distintos sustrados, basados en predicciones de FBA, coinciden con resultados experimentales (Edwards, 2001). 

Debido a que hay información disponible de organismos unicelulares, muchas de sus redes metabólicas son objeto de estudio. No obstante, es de esperarse que en un futuro no muy lejano se estudien redes de organismos multicelulares. Por otra parte, modelos tales como *ME-Model* que integran información tanto metabólica como de expresión, resultan ser una primera aproximación para lograr conjuntar varios niveles de conocimiento y así llegar a modelar por completo una célula lo más fielmente posible.
